```
#Instalação das bibliotecas
import requests
from bs4 import BeautifulSoup
```

```
link   = "https://www.br.undp.org/content/brazil/pt/home/idh0/rankings/idh-global.html"

page = requests.get(link).text

print(page)
```

```
soup = BeautifulSoup(page,'html5lib')
```

```
dataset = []
for tr in soup.select('.parbase.section.text .tableizer-table tbody tr'):
  tds = tr.select('td')
  if len(tds) >= 3 :
    pais = tds[1].get_text().replace('\n','')
    IDH = tds[2].get_text().replace('\n','')
    print(pais, IDH)
    dataset.append((pais,IDH))
```

```
#salvando como arq csv

with open('IDH_paises.csv','w') as f:
  f.write('"pais","IDH"\n')
  for linha in dataset:
    #f.write('"&s","&s"\n' & linha[0],linha[1])
    f.write('"{}","{}"\n' .format(linha[0],linha[1]))
```
